1. What is the difference between multithreading and multiprocessing?

Answer

* Multithreading : Runs multiple threads within the same process. They share
memory and resources, making it lightweight and good for tasks like I/O operations or UI updates.

* Multiprocessing : Runs multiple separate processes, each with its own memory space. It’s heavier but better for CPU-bound tasks like data crunching or parallel computations.

Quick analogy: Multithreading is like several people working in the same room (shared space). Multiprocessing is like each person working in their own office (isolated space).



2. What are the challenges associated with memory management in Python?

Answer

* Garbage collection delays : Python uses automatic garbage collection, but sometimes unused memory isn’t freed quickly, leading to memory bloat.

* Reference cycles : Objects that reference each other can create cycles that are hard for the garbage collector to clean up.

* Global Interpreter Lock (GIL) : Limits true parallel execution in threads, which can affect memory efficiency in multi-threaded programs.

* Memory leaks : Poorly written code or third-party libraries can accidentally hold onto memory longer than needed.

* Large object overhead : Python objects have extra metadata, which can make them heavier than in lower-level languages.



3. Write a Python program that logs an error message to a log file when a division by zero exception occurs.

Answer


```python
import logging

# Set up logging
logging.basicConfig(filename='error.log', level=logging.ERROR)

try:
    result = 10 / 0
except ZeroDivisionError as e:
    logging.error("Division by zero error occurred: %s", e)
```

5. Write a program that handles both IndexError and KeyError using a
try-except block.

Answer

```python
my_list = [1, 2, 3]
my_dict = {'a': 10, 'b': 20}

try:
    print(my_list[5])
    print(my_dict['z'])
except IndexError:
    print("Oops! List index out of range.")
except KeyError:
    print("Oops! Key not found in dictionary.")
```

6. What are the differences between NumPy arrays and Python lists?

Answer

NumPy Arrays vs Python Lists:

*Speed: NumPy arrays are faster for large data operations.

*Memory: NumPy uses less memory than Python lists.

*Data type: NumPy arrays store elements of the same type; lists can mix types.

*Math operations: NumPy supports direct math (like array1 + array2); lists need loops.

*Shape: NumPy handles multi-dimensional data easily; lists need nesting.

*Functions: NumPy has tons of built-in functions for stats, algebra, etc.; lists are basic.

7. Explain the difference between apply() and map() in Pandas.

Answer

map() — For Series Only
* Works only on Series (single column).

* Applies a function element-wise to each value.

* Accepts functions, dictionaries, or Series for mapping.

* Ideal for simple transformations or value replacements.

* Cannot be used on DataFrames.

* Internally optimized for speed with basic operations.

apply() — For Series & DataFrames
* Works on both Series and DataFrames.

* Can apply functions element-wise, row-wise, or column-wise.

* Accepts any function, including custom logic.

* Ideal for complex operations, especially across rows or multiple columns.

* Supports axis=0 (column-wise) and axis=1 (row-wise) for DataFrames.

* Can return scalars, Series, or even DataFrames depending on the function.

8.  Create a histogram using Seaborn to visualize a distribution.

Answer

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [55, 60, 65, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98, 100]

sns.histplot(data, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of Exam Scores')
plt.show()
```

10.  Calculate the correlation matrix using Seaborn and visualize it with a
heatmap.

Answer

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Student score.csv')
corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
